```python
import gradio as gr

def predict_genre(plot):
    plot_vector = tfidf.transform([plot])
    prediction = model.predict(plot_vector)
    return prediction[0]

demo = gr.Interface(
    fn=predict_genre,
    inputs=gr.Textbox(lines=5, label="Movie Plot"),
    outputs=gr.Textbox(label="Predicted Genre"),
    title="🎬 Movie Genre Classification",
    description="Enter a movie plot and predict its genre."
)

demo.launch()
```
